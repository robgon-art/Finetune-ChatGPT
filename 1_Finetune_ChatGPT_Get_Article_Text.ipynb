````python
import glob
html_files = glob.glob('/content/robgon_articles_html/*.html')
print(len(html_files))
html_files.sort()
for h in html_files:
  print(h)
````

````python
nickname_list = []
with open('/content/nicknames.txt') as f:
  for line in f:
    line = line.strip()
    nickname_list.append(line)

print(len(nickname_list))
````

````python
page_list = []
with open('/content/urls.txt') as f:
  for line, nickname in zip(f, nickname_list):
    line = line.strip()
    page = line.split('/')[-1]
    page_list.append((page, nickname, line))

print(len(page_list))
print(page_list)
page_list.sort()
for pl in page_list:
  print(pl)
````

````python
from bs4 import BeautifulSoup, Comment
from datetime import datetime
import re

# Define the styles for markdown
md_styles = {
    'h1': '# ',
    'h2': '## ',
    'pre': '',
    'blockquote': '> ',
    'p': '',
}

def get_section_filename(base_filename, section_number, ext):
    return f"{base_filename}_{section_number}.{ext}"

def write_to_txt_file(i, pair, txt_file):
  line = pair[1]
  txt_file.write(line + '\n')
  if i == 3:
    txt_file.write(date_text + '\n')
    txt_file.write(nickname + ' - ')
    txt_file.write(url + '\n')

def write_to_md_file(i, pair, md_file):
  style = pair[0]
  line = pair[1]

  if i == 3:
    md_file.write('</br>' + date_text + '</br>\n')
    md_file.write(nickname + ' - ')
    md_file.write(url + '</br></br>\n')
  elif style == 'pre':
    md_file.write("```\n" + line + "\n```\n")
  else:
    prefix = md_styles.get(style, '')
    md_file.write(prefix + line + '\n')
    if re.match(r'\[\d+\]\s+.*', line):
      md_file.write('\n')
  if style == 'blockquote' and (i+1 < len(filtered_texts)) and filtered_texts[i+1][0] != 'blockquote':
    md_file.write('\n')  # Blank line to terminate the blockquote context.
    md_file.write("\n")

for input_file, page in zip(html_files, page_list):
    print(input_file)

    # Load the HTML content
    with open(input_file, 'r', encoding='utf-8') as file:
        html_content = file.read()

    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Define a function to check if an element is visible
    def is_visible(element):
        if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
            return False
        if isinstance(element, Comment):
            return False
        return True

    # Extract the date text
    date_element = soup.find('span', {'data-testid': 'storyPublishDate'})
    if date_element:
        date_text = date_element.get_text().strip()

    # Find title, subtitle, headings, and paragraph elements
    elements_to_extract = ['title', 'h1', 'h2', 'p', 'pre']
    text_elements = []
    for element in soup.find_all(True): # iterate through all elements
        if element.name in elements_to_extract:
            if element.find_parent('blockquote'):
              # print("blockquote", element)
              text_elements.append(("blockquote", element))
            else:
              # print(element.name, element)
              text_elements.append((element.name, element))

    visible_texts = []
    for pair in text_elements:
        element = pair[1]
        if is_visible(element):
            text = element.get_text(separator=' ').strip()
            if text == "Some rights reserved" or text.startswith("To get unlimited access"):
              # print(text)
              break
            visible_texts.append((pair[0], text))

    filtered_texts = []
    for i, line in enumerate(visible_texts):
        # print(i, line)
        if i not in [0,5,6,7,8,9]:
          filtered_texts.append(line)

    if len(date_text) <= 6:
        date_text += ", 2023"

    date_obj = datetime.strptime(date_text, '%b %d, %Y')
    formatted_date = date_obj.strftime('%Y-%m-%d')

    output_file_txt = input_file.replace("html", "txt")
    output_file_txt = output_file_txt.replace("/robgon_articles_txt/", "/robgon_articles_txt/" + formatted_date + "_")

    output_file_md = input_file.replace("html", "md")
    output_file_md = output_file_md.replace("/robgon_articles_md/", "/robgon_articles_md/" + formatted_date + "_")

    nickname = page[1]
    url = page[2]
    print(output_file_txt)
    print(output_file_md)
    print(nickname, '-', url)
    print()

    with open(output_file_txt, 'w', encoding='utf-8') as txt_file, open(output_file_md, 'w', encoding='utf-8') as md_file:
      for i, pair in enumerate(filtered_texts):
        write_to_txt_file(i, pair, txt_file)
        write_to_md_file(i, pair, md_file)

    quarter_length = len(filtered_texts) // 4

    sections = [
        filtered_texts[:quarter_length],
        filtered_texts[quarter_length:2*quarter_length],
        filtered_texts[2*quarter_length:3*quarter_length],
        filtered_texts[3*quarter_length:]
    ]

    for section_number, section in enumerate(sections, 1):
        output_file_txt_section = output_file_txt.replace("articles", "article_parts")
        output_file_txt_section = get_section_filename(output_file_txt_section.rstrip('.txt'), section_number, 'txt')
        output_file_md_section = output_file_md.replace("articles", "article_parts")
        output_file_md_section = get_section_filename(output_file_md_section.rstrip('.md'), section_number, 'md')

        with open(output_file_txt_section, 'w', encoding='utf-8') as txt_file, open(output_file_md_section, 'w', encoding='utf-8') as md_file:
            for i, pair in enumerate(section):
                write_to_txt_file(i, pair, txt_file)
                write_to_md_file(i, pair, md_file)
````
